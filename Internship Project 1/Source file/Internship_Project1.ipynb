#Machine Learning Part

```python
#scikit learn
#algorithm 1 - importing Library
from sklearn import linear_model
```

```python
#algorithm 2 - create a reference object to the linear model module
reg=linear_model.LinearRegression()
```

```python
#algorithm 3 - we have to train the data
reg.fit(X,Y)
```

```python
#algorithm 4 - Predict the output / testing the data
reg.predict([[2500]])
```

```python
#algorithm 5 - plotting the ML best fit line
Ytest=reg.predict(X)
Ytest
```

```python
plt.scatter(X,Y)
plt.plot(X,Y)
```

```python
plt.scatter(X,Y)
plt.plot(X,Ytest)
```

```python
#accuracy
reg.score(X,Y)
```

#Understanding the ML

```python
#Y=mX+b
in1=2500
m=reg.coef_
b=reg.intercept_
```

```python
Y=m*in1+b
Y
```
